```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')  
%matplotlib inline
pd.set_option('display.max_rows', 100)
```

```python
nutri_data=pd.read_excel('mcdonalds_india_menu.xlsx')
nutri_data.head()
```

# Univariate plotting

```python
nutri_data['Category'].value_counts().plot(kind='bar',figsize=(15,10),fontsize=15,yticks=np.arange(0,19,2),)


plt.xlabel('Category',fontsize=20)
plt.ylabel('No. of products',fontsize=20)
plt.title('Bar Graph showing products available in each Category',fontsize=23)
```

```python
nutri_data['Energy'].plot(kind='line',figsize=(15,10),linestyle='-.',fontsize=15)
plt.xlabel('Energy (Kcal)', fontsize=20)
plt.ylabel('Index of Item',fontsize=20)
plt.title('Line graph depicting Energy for each item on menu',fontsize=23)
```

```python
nutri_data['Total_Fat'].plot(kind='area',figsize=(15,8),cmap='magma',fontsize=15,xticks=np.arange(0,91,10))

plt.xlabel('Index',fontsize=20)
plt.ylabel('Total_Fat',fontsize=20)
plt.title('Area chart showing the variation of Total Fat across different items',fontsize=23)
```

```python
print('The categories with the highest total fats are')
nutri_data[15:28]['Category'].value_counts()
```

```python
print('The Item with the highest fat content is ',nutri_data.iloc[nutri_data.iloc[0:28]['Total_Fat'].argmax()]['Item'], ' and it is from the category ', nutri_data.iloc[nutri_data.iloc[0:28]['Total_Fat'].argmax()]['Category'])
```

```python
bins = [0, 50, 100, 150, 200, 250, 300, 350, 400, 450, 500]
nutri_data['Serve_Size'].plot(kind='hist', bins = bins, figsize=(15, 7), color='brown', yticks=np.arange(0, 22, 3), xticks=bins)
```

```python
nutri_data['Serve_Size'].plot(kind='kde',figsize=(15,10),fontsize=15)
```

```python
space=np.ones(11)/20

nutri_data['Category'].value_counts().plot(kind='pie',explode=space,figsize=(12,12),cmap='magma',fontsize=13, startangle=160,
                                           legend=True, autopct='%3.2f%%', wedgeprops=dict(width=0.2),shadow=True)

plt.ylabel('Category',fontsize=15)
plt.title('Donut Plot showing the proportion of each Category value',fontsize=15)
```
